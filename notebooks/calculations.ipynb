```python
from sympy import MatrixSymbol, Matrix, solve, Eq
```

```python
X = MatrixSymbol('X', 10, 3)
b= MatrixSymbol('beta', 3, 1)
y = MatrixSymbol('y', 10, 1)
```

```python
ans = ((y - X*b).T*(y - X*b)).diff(b)
solve(ans.expand()[0,0], b)
```

```python
from sympy import Matrix, symbols, diff, solve, MatrixSymbol, eye

# Define the dimensions
n, p = 10, 3

# Create symbolic matrices
X = Matrix(MatrixSymbol('X', n, p))
beta = Matrix(symbols('beta1:' + str(p+1)))
y = Matrix(MatrixSymbol('y', n, 1))

# Calculate the objective function
obj_func = (y - X*beta).T * (y - X*beta)

# Take the derivative with respect to beta
gradient = diff(obj_func[0, 0], beta)

# Set up the normal equations
normal_eqs = gradient.T

# Solve the normal equations
solution = solve(normal_eqs, beta)

print("Solution:")
print(solution)

# Alternatively, we can derive the closed-form solution
X_T = X.T
closed_form = (X_T * X).inv() * X_T * y

print("\nClosed-form solution:")
print(closed_form)
```
